```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
df = pd.read_csv("titanic.csv")
```

```python
print("Shape of Dataset:", df.shape)
```

```python
df["Survived"].value_counts()
```

```python
df["Survived"].value_counts(normalize=True) * 100
```

```python
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Survival by Gender")
plt.savefig("images/survival_gender.png")
plt.show()
```

```python
sns.countplot(x="Pclass", hue="Survived", data=df)

plt.title("Survival by Passenger Class")

plt.savefig("images/survival_class.png")

plt.show()
```

```python
bins = [0,12,19,35,60,100]
labels = ["Child","Teen","Young Adult","Adult","Senior"]

df["Age Group"] = pd.cut(df["Age"], bins=bins, labels=labels)
```

```python
sns.countplot(x="Age Group", hue="Survived", data=df)

plt.title("Survival by Age Group")

plt.savefig("images/survival_age_group.png")

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap="coolwarm")

plt.title("Correlation Heatmap")

plt.savefig("images/heatmap.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Age"], bins=30)

plt.title("Age Distribution")

plt.savefig("images/age_distribution.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Fare"], bins=30)

plt.title("Fare Distribution")

plt.savefig("images/fare_distribution.png")

plt.show()
```

```python
sns.countplot(x="Pclass", data=df)

plt.title("Passenger Class Distribution")

plt.savefig("images/passenger_class_distribution.png")

plt.show()
```

# Key Findings

1. Female passengers had a much higher survival rate than male passengers.

2. First-class passengers had the highest survival rate.

3. Most passengers belonged to the Young Adult age group.

4. Passengers who paid higher fares were more likely to survive.

5. The Cabin column contains many missing values, while the Age column has some missing values that should be handled before further analysis.
